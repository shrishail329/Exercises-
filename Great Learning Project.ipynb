4. Write a function with name â€œGoal_diff_countâ€ which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference.
 
 Goal_diff_count = GoalsFor - GoalsAgainst

```python
## Convert the str/object to number 

df['GoalsFor'] = pd.to_numeric(df['GoalsFor'],errors='coerce')
df['GoalsAgainst'] = pd.to_numeric(df['GoalsAgainst'],errors='coerce')

df['Goalsdifference'] = df['GoalsFor'] - df['GoalsAgainst']

df[['Goalsdifference']]
```

Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage

```python
## Coverted the str/object to numerical values then performed the problem

df['GamesWon'] = pd.to_numeric(df['GamesWon'],errors='coerce')

df['GamesPlayed'] = pd.to_numeric(df['GamesPlayed'],errors='coerce')

## Calculated the winning percentage 

df['Winning Percent'] = df['GamesWon']/df['GamesPlayed']*100

# print the Winning percentage 

df[['Winning Percent']]
```
